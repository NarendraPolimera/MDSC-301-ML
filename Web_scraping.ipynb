```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm.notebook import tqdm_notebook
```

```python
from bs4 import BeautifulSoup
def scrape(text):
  doc = BeautifulSoup(text, "html.parser")
  res = doc.findChildren('span', {'data-hook':'review-body'})
  # res = [x.findChild('span').text for x in res]
  res = [x.text for x in res]
  return res
```

```python
#the website URL
reviews = []
url_link = "https://www.amazon.in/Noise-Advanced-Bluetooth-Detection-Battery/product-reviews/B0B5LW5DFQ/ref=cm_cr_arp_d_paging_btm_next_73?ie=UTF8&reviewerType=all_reviews&pageNumber="
for i in range(500):
  page = requests.get(url_link + str(i))
  print("Page Status: ", i, " ", page.status_code)
  i+=1
  result = scrape(page.text)
  reviews.extend(result)
```
